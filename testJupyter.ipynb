```python
# collatz iterations using recursive function
```

```python
def collatz(x,n=0):
    """returns number of iterations to arrive to 1
    for the first time
    """
    if type(x)!=int or x<1:
        raise TypeError('Enter an integer')
    else:
        if x==1:
            return n   
        if x%2==0:
            x=x/2
            if x==1:
                return n+1
            else:
                return collatz(x,n+1)
        else:
            return collatz(3*x+1,n+1)
```

```python
# some plots of the collatz number of iterations required to arrive to 1
```

```python
bounces=[]
for i in range(1,10000):
    bounces.append(collatz(i))
plt.plot(bounces,'x',alpha=0.9)
plt.title('some points')
plt.show()
```
